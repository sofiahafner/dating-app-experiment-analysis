```python
import pandas as pd
import matplotlib.pyplot as plt
import textwrap
import ast
import glob
import os
import numpy as np
```

```python

base_path = "/Users/franziskahafner/Desktop/oii/trinity/thesis/experiment_data/prolific_data/Control/"
batch_folders = glob.glob(os.path.join(base_path, "batch*"))

batch_df = pd.DataFrame()
game_df = pd.DataFrame()
global_df = pd.DataFrame()
player_df_org = pd.DataFrame()
round_df = pd.DataFrame()
stage_df = pd.DataFrame()
playerGame_df = pd.DataFrame()
playerRound_df = pd.DataFrame()
playerStage_df = pd.DataFrame()
demographic_df = pd.DataFrame()

for batch_folder in batch_folders:
    date_folders = [f for f in glob.glob(os.path.join(batch_folder, "dating-app-experiment-*")) if os.path.isdir(f)]
    if not date_folders:
        continue
    date_folder = date_folders[0]
    
    batch_df = pd.concat([batch_df, pd.read_csv(os.path.join(date_folder, "batch.csv"))])
    game_df = pd.concat([game_df, pd.read_csv(os.path.join(date_folder, "game.csv"))])
    global_df = pd.concat([global_df, pd.read_csv(os.path.join(date_folder, "global.csv"))])
    player_df_org = pd.concat([player_df_org, pd.read_csv(os.path.join(date_folder, "player.csv"))])
    round_df = pd.concat([round_df, pd.read_csv(os.path.join(date_folder, "round.csv"))])
    stage_df = pd.concat([stage_df, pd.read_csv(os.path.join(date_folder, "stage.csv"))])
    playerGame_df = pd.concat([playerGame_df, pd.read_csv(os.path.join(date_folder, "playerGame.csv"))])
    playerRound_df = pd.concat([playerRound_df, pd.read_csv(os.path.join(date_folder, "playerRound.csv"))])
    playerStage_df = pd.concat([playerStage_df, pd.read_csv(os.path.join(date_folder, "playerStage.csv"))])
    
    demographic_files = glob.glob(os.path.join(batch_folder, "prolific_export_*.csv"))
    if demographic_files:
        demographic_file = demographic_files[0]
        demographic_df = pd.concat([demographic_df, pd.read_csv(demographic_file)])

batch_df.reset_index(drop=True, inplace=True)
game_df.reset_index(drop=True, inplace=True)
global_df.reset_index(drop=True, inplace=True)
player_df_org.reset_index(drop=True, inplace=True)
round_df.reset_index(drop=True, inplace=True)
stage_df.reset_index(drop=True, inplace=True)
playerGame_df.reset_index(drop=True, inplace=True)
playerRound_df.reset_index(drop=True, inplace=True)
playerStage_df.reset_index(drop=True, inplace=True)
demographic_df.reset_index(drop=True, inplace=True)
```

```python
def extract_feedback(survey):
    if pd.isna(survey):
        return np.nan
    try:
        feedback_dict = json.loads(survey)
        feedback = feedback_dict.get('feedback', '')
        return feedback if feedback else np.nan
    except json.JSONDecodeError:
        return np.nan

playerGame_df['feedback'] = playerGame_df['exitSurvey'].apply(extract_feedback)
```

## Likes in Past Rounds to inform Recommendations:

The json from this is then included in the treatment group experiments, to provide information on which other profiles have liked the current profile:

```python
import json
from json import loads, dumps
import ast 

def str_to_list(list_str):
    try:
        return ast.literal_eval(list_str)
    except ValueError:
        return [] 

playerGame_df['likedProfiles'] = playerGame_df['likedProfiles'].apply(str_to_list)
playerGame_df['dislikedProfiles'] = playerGame_df['dislikedProfiles'].apply(str_to_list)

records = []
for index, row in playerGame_df.iterrows():
    for liked, disliked in zip(row['likedProfiles'], row['dislikedProfiles']):
        record = {
            "participantID": row['id'],
            "chosenProfileID": row['chosenProfile'],
            "recommendedProfiles": [liked, disliked],
            "likedProfile": liked,
            "dislikedProfile": disliked
        }
        records.append(record)

expanded_df = pd.DataFrame(records)
json_output = expanded_df.to_json(orient='records')
parsed_json = json.loads(json_output)

with open('/Users/franziskahafner/Desktop/oii/trinity/thesis/experiment/dating-app-experiment/server/data/likesPastRounds.json', 'w') as f:
    json.dump(parsed_json, f, indent=2)
```

```python
relevant_prolific_data = demographic_df[['Participant id', 'Status', 'Started at', 'Completed at', 'Time taken', 'Dating apps', 'Gender', 'Age', 'Sex',
       'Ethnicity simplified', 'Country of birth', 'Country of residence',
       'Nationality', 'Language', 'Student status', 'Employment status']]
```

```python
player_df = playerGame_df[['id', 'chosenProfile', 'playerID',  "participantIdentifier", 'recAlgorithm', 'ended', 'feedback']]
player_df = player_df.merge(relevant_prolific_data, left_on='participantIdentifier', right_on='Participant id', how='left')
player_df.drop(columns=['Participant id'], inplace=True)
def clean_identifier(identifier):
    return identifier.split('@')[0]
player_df['participantIdentifier'] = player_df['participantIdentifier'].apply(clean_identifier)
```
