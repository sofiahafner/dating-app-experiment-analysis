```python
def build_preference_dictionary(liked_profiles, disliked_profiles, character_df_encoded):
    features = [
        'young', 'hair_short', 'skin_dark', 'has_facial_hair', 'has_accessories', 'job_type_higher_ed', 
        'face_positive', "main_hobby_Arts", "main_hobby_Music", "main_hobby_Sport", "main_hobby_Travel"
    ]

    preferences = {feature: 0 for feature in features}
    comparisons = {feature: 0 for feature in features}
    sum_squares = {feature: 0 for feature in features}
    total_observations = 0

    for index, liked_profile in enumerate(liked_profiles):
        if index < len(disliked_profiles):
            liked_features = character_df_encoded.loc[liked_profile]
            disliked_features = character_df_encoded.loc[disliked_profiles[index]]

            for feature in features:
                if liked_features[feature] and not disliked_features[feature]:
                    preferences[feature] += 1
                    sum_squares[feature] += 1 ** 2
                    comparisons[feature] += 1
                elif not liked_features[feature] and disliked_features[feature]:
                    preferences[feature] -= 1
                    sum_squares[feature] += (-1) ** 2
                    comparisons[feature] += 1
                if liked_features[feature] or disliked_features[feature]:
                    total_observations += 1

    normalized_preferences = {}
    for feature in features:
        if comparisons[feature] != 0:
            mean = preferences[feature] / comparisons[feature]
            variance = (sum_squares[feature] / comparisons[feature]) - (mean ** 2)
            normalized_preferences[feature] = {
                'mean': mean * 100,
                'variance': variance * 100,
                'n': comparisons[feature],
            }
        else:
            normalized_preferences[feature] = {
                'mean': 0,
                'variance': 0,
                'n': 0,
            }

    return normalized_preferences
```

### Stability:

```python
character_df_encoded = pd.get_dummies(character_df, columns=['main_hobby'])

all_normalized_preferences = []

features = [
    'young', 'hair_short', 'skin_dark', 'has_facial_hair', 'has_accessories', 'job_type_higher_ed', 'face_positive'
] + [col for col in character_df_encoded.columns if col not in ['profile_ID']]

for current_round in range(1, 51):
    current_round_data = SwipeDecision_df[SwipeDecision_df['roundsPlayed'] <= current_round]

    preferences = {}
    comparisons = {}

    for _, row in current_round_data.iterrows():
        playerProfile = row['playerProfile']
        dislikedProfile = row['dislikedProfile']
        likedProfile = row['likedProfile']

        if playerProfile not in preferences:
            preferences[playerProfile] = {feature: 0 for feature in features}
            comparisons[playerProfile] = {feature: 0 for feature in features}

        liked_features = character_df_encoded[character_df_encoded['profile_ID'] == likedProfile].iloc[0].to_dict()
        disliked_features = character_df_encoded[character_df_encoded['profile_ID'] == dislikedProfile].iloc[0].to_dict()

        for feature in features:
            if liked_features[feature] and not disliked_features[feature]:
                preferences[playerProfile][feature] += 1
                comparisons[playerProfile][feature] += 1
            elif not liked_features[feature] and disliked_features[feature]:
                preferences[playerProfile][feature] -= 1
                comparisons[playerProfile][feature] += 1

    normalized_preferences = {}
    for playerProfile, prefs in preferences.items():
        normalized_preferences[playerProfile] = {}
        for feature, value in prefs.items():
            if comparisons[playerProfile][feature] != 0:
                normalized_preferences[playerProfile][feature] = (value / comparisons[playerProfile][feature]) * 100
            else:
                normalized_preferences[playerProfile][feature] = 100

    all_normalized_preferences.append(normalized_preferences)

```

```python
average_differences = {feature: [] for feature in features}

for round_index in range(1, 50):
    current_round_prefs = all_normalized_preferences[round_index]
    previous_round_prefs = all_normalized_preferences[round_index - 1]

    differences = {feature: [] for feature in features}
    for playerProfile in current_round_prefs:
        for feature in features:
            current_pref = current_round_prefs[playerProfile].get(feature, 0)
            previous_pref = previous_round_prefs[playerProfile].get(feature, 0)
            difference = current_pref - previous_pref
            differences[feature].append(difference)

    for feature in features:
        if differences[feature]:
            average_difference = sum(differences[feature]) / len(differences[feature])
        else:
            average_difference = 0
        average_differences[feature].append(average_difference)
```

```python
differences_list = copy.deepcopy(all_normalized_preferences)

for i in range(1, len(all_normalized_preferences)):
    for key in all_normalized_preferences[i]:
        for feature in all_normalized_preferences[i][key]:
            previous_value = all_normalized_preferences[i-1][key][feature]
            current_value = all_normalized_preferences[i][key][feature]
            differences_list[i][key][feature] = abs(current_value - previous_value)

features = list(all_normalized_preferences[0][list(all_normalized_preferences[0].keys())[0]].keys())
average_differences = {feature: [] for feature in features}

for feature in features:
    for i in range(1, len(differences_list)):
        total_difference = 0
        count = 0
        for key in differences_list[i]:
            if feature in differences_list[i][key]:
                total_difference += differences_list[i][key][feature]
                count += 1
        average_differences[feature].append(total_difference / count if count > 0 else 0)
```

```python
average_differences_df = pd.DataFrame(average_differences)

num_features = len(average_differences_df.columns)

num_cols = 3
num_rows = (num_features // num_cols) + (num_features % num_cols > 0)

fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, num_rows * 2.5), dpi=350)
axes = axes.flatten()

for i, feature in enumerate(average_differences_df.columns):
    axes[i].plot(average_differences_df[feature], marker='o', linestyle='-')
    axes[i].set_title(feature)
    axes[i].set_xlabel('Rounds')
    axes[i].set_ylabel('Average Difference')
    axes[i].set_ylim(0, 50)
    axes[i].grid(True)

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
